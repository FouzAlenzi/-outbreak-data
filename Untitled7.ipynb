```python
import pandas as pd
import numpy as np

df = pd.read_csv("NORS_20250803 (3).csv", low_memory=False)

df.head()
```

```python
df.info()

df.isna().sum()
```

```python
df = df.dropna(axis=1, how="all")

for col in df.columns:
    if df[col].nunique() <= 1:
        df = df.drop(columns=[col])

for col in df.columns:
    if "date" in col.lower():
        df[col] = pd.to_datetime(df[col], errors="coerce")

df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)

df = df.drop_duplicates()

df.shape
```

```python
numeric_cols = ["Hospitalizations", "Info On Hospitalizations", "Deaths", "Info On Deaths"]

df[numeric_cols] = df[numeric_cols].fillna(0).astype(int)

df[numeric_cols].info()
```

```python
import matplotlib.pyplot as plt

top_states = df.groupby("State")["Illnesses"].sum().sort_values(ascending=False).head(5)

plt.figure(figsize=(8,6))
top_states.plot(kind="bar", color="skyblue", edgecolor="black")

plt.title("Top 5 States by Illnesses")
plt.xlabel("State")
plt.ylabel("Total Illnesses")
plt.xticks(rotation=45)
plt.show()
```

```python
illnesses_by_year = df.groupby("Year")["Illnesses"].sum()

illnesses_last10 = illnesses_by_year.tail(10)

import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
illnesses_last10.plot(kind="line", marker="o", color="darkorange")

plt.title("Trend of Illnesses (Last 10 Years)")
plt.xlabel("Year")
plt.ylabel("Total Illnesses")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

```python
# نجمع التنويم حسب الـ Setting

hosp_by_setting = df.groupby("Setting")["Hospitalizations"].sum().sort_values(ascending=False).head(5)

# Bar chart
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
hosp_by_setting.plot(kind="barh", color="teal", edgecolor="black")

plt.title("Top 5 Settings by Total Hospitalizations")
plt.xlabel(" Number of Hospitalizations")
plt.ylabel(" Setting")
plt.xticks(rotation=45)
plt.show()
```

```python
top_deaths_by_cause = df.groupby("Etiology")["Deaths"].sum().sort_values(ascending=False).head(5)

plt.figure(figsize=(8,6))
top_deaths_by_cause.plot(kind="bar", color="crimson", edgecolor="black")

plt.title("Top 5 Etiologies by Total Deaths")
plt.xlabel("Etiology")
plt.ylabel("Number of Deaths")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
outbreaks_per_year = df.groupby("Year").size()
last_10_years = outbreaks_per_year.sort_index().tail(10) 

plt.figure(figsize=(10,6))
last_10_years.plot(kind="line", marker="o", color="navy")
plt.title("Number of Outbreaks in the Last 10 Years", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Number of Outbreaks")
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()
```

```python
recent_df = df[df["Year"] >= (df["Year"].max() - 10)]

top_foods = (
    recent_df["Food Vehicle"]
    .value_counts()
    .head(5)
)

plt.figure(figsize=(8,5))
top_foods.plot(kind="bar", color="teal")
plt.title("Top 5 Food Vehicles in the Last 10 Years", fontsize=14)
plt.xlabel("Food")
plt.ylabel("Number of Outbreaks")
plt.xticks(rotation=30, ha="right")
plt.show()
```

```python
mode_counts = df["Primary Mode"].value_counts().head(5)

# Bar chart
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
mode_counts.plot(kind="bar", color="mediumpurple", edgecolor="black")

plt.title("Distribution of Primary Modes")
plt.xlabel("Mode of Transmission")
plt.ylabel("Count of Outbreaks")
plt.xticks(rotation=45)
plt.show()

# Pie chart
plt.figure(figsize=(7,7))
mode_counts.plot(kind="pie", autopct='%1.1f%%', startangle=140, cmap="Pastel1")
plt.title("Primary Modes Distribution (Top 5)")
plt.ylabel("")
plt.show()
```
